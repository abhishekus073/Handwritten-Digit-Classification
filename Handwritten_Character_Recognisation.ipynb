```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('A_Z Handwritten Data.csv').astype('float32')
data.head(10)
```

```python
X = data.drop('0',axis = 1)
y = data['0']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

x_train = np.reshape(x_train.values, (x_train.shape[0], 28,28))
x_test = np.reshape(x_test.values, (x_test.shape[0], 28,28))

print("Shape of Training data: ", x_train.shape)
print("Shape of Testing data: ", x_test.shape)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

# Shuffle training data
shuffle_data = shuffle(x_train)

# Create 3x3 grid
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.flatten()

for i in range(9):
    # Apply threshold to the shuffled image
    _, shu = cv2.threshold(shuffle_data[i], 30, 200, cv2.THRESH_BINARY)

    # Display thresholded image
    axes[i].imshow(shu.reshape(28, 28), cmap="gray")
    axes[i].axis("off")

plt.show()
```

```python
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1],x_train.shape[2],1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2],1)

print("New shape of training data: ", x_train.shape)
print("New shape of testing data: ", x_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python
from keras.utils import to_categorical

y_training = to_categorical(y_train, num_classes=26)
y_testing = to_categorical(y_test, num_classes=26)


print("New shape of training labels: ", y_training.shape)
print("New shape of testing labels: ", y_testing.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense

# Define CNN model
model = Sequential()

# 1st Convolution + Pooling
model.add(Conv2D(64, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPool2D(pool_size=(2, 2)))

# 2nd Convolution + Pooling
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))

# 3rd Convolution + Pooling
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))

# Flatten for fully connected layers
model.add(Flatten())

# Dense layers
model.add(Dense(128, activation="relu"))
model.add(Dense(256, activation="relu"))

# Output layer (26 classes for A–Z alphabets)
model.add(Dense(26, activation="softmax"))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()
```

```python
model.compile(optimizer = Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.save(r'handwritten_character_recog_model.h5')
```

```python
words = {0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J',10:'K',11:'L',12:'M',13:'N',14:'O',15:'P',16:'Q',17:'R',18:'S',19:'T',20:'U',21:'V',22:'W',23:'X', 24:'Y',25:'Z'}

fig, axes = plt.subplots(3,3, figsize=(8,9))
axes = axes.flatten()

for i,ax in enumerate(axes):
    image = np.reshape(x_test[i], (28,28))
    ax.imshow(image, cmap="Greys")
    
    pred = words[np.argmax(y_testing[i])]
    ax.set_title("Prediction: "+pred)
    ax.grid()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
history = model.fit(x_train, y_training, 
                    epochs=5, 
                    validation_data=(x_test, y_testing),
                    batch_size=32)
# Final training accuracy
final_train_acc = history.history['accuracy'][-1]

# Final validation accuracy
final_val_acc = history.history['val_accuracy'][-1]

print(f"Final Training Accuracy: {final_train_acc * 100:.2f}%")
print(f"Final Validation Accuracy: {final_val_acc * 100:.2f}%")

```
